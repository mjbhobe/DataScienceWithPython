```python
# BoW implementation
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
file_name = "sherlock_holmes.txt"
corpus = []
skip_lines = 31

with open(file_name, "r") as f:
    data = f.readlines()
    for line in data:
        line = line.strip().replace("\n", " ")
        if len(line) > 0:
            # skip blank lines
            corpus.append(line)

# print(corpus)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
corpus_tfidf = vectorizer.fit_transform(corpus)
# what is the vocab size?
print(f"Vocab size: {len(vectorizer.vocabulary_.keys())}")
print(f"Document matrix shape: {corpus_tfidf.shape}")
df = pd.DataFrame(np.round(corpus_tfidf.toarray(), 2))
df.columns = vectorizer.get_feature_names_out()
print(df.shape)
print(df.head())
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC

labels = np.random.randint(0, 3, (len(df), 1)).ravel()
clf = SVC()
clf.fit(df.to_numpy(), labels)
```

```python
from sklearn.metrics import accuracy_score

pred = clf.predict(df.to_numpy())
print(accuracy_score(labels, pred))
```
