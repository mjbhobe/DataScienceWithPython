## Exploring your dataset with `pandas`
In this notebook we walk through teh basics of exploring structured datasets (such as CSV files) using the `pandas` library.

```python
import os
import pathlib
import shutil
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# pandas tweaks
pd.set_option("display.max_columns", None)  # show all cols
pd.set_option("display.precision", 2)   # floating point precision = 2
```

```python
def download_data() -> str:
    # download dataset to local directory
    url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-elo/nbaallelo.csv"
    data_dir = pathlib.Path(os.getcwd()) / "data"

    # download file, if not already
    data_file_path = data_dir / "nbaallelo.csv"
    response = requests.get(url)
    response.encoding = "utf-8"

    if not data_file_path.exists():
        response = requests.get(url)
        response.encoding = "utf-8"
        if response.status_code == 200:
            # we were able to download the file, move it to data folder
            with open(data_file_path, "wb") as f:
                f.write(response.content)
        else:
            print(f"FATAL: unable to download file from {url}")
    else:
        return str(data_file_path)
```

```python
df = pd.read_csv(download_data())
df.tail() # show last 5 rows
```
